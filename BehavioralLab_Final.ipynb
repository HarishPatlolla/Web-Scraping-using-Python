# Scraping of donations using the URLs. CEO can have more than 200 donations ( 4*50 per page) 

```python
from selenium import webdriver
from bs4 imporat BeautifulSoup
import pandas as pd

#Configure chrome driver#
driver = webdriver.Chrome("chromedriver")

Notmapped_Donations=[]


for i in range(len(Not_Mapped_donations_pd_list1)):
    url= 'https://www.opensecrets.org/donor-lookup/results?name='+Not_Mapped_donations_pd_list1[i]
    print(url)
    for j in range(3):
        url_update=url+'&page='+str(j)
        print(url_update)
        driver.get(url_update)
        content = driver.page_source
        soup = BeautifulSoup(content)
        table_rows = soup.find_all('tr')

        for tr in table_rows:
            td = tr.find_all('td')
            row = [tr.text for tr in td]
            if row:
                Notmapped_Donations.append(row)
```

# Removing the duplicates and records with null donation amount

```python
#Combination of str[0] & str[2] of scraped data with multiple combinations of master set
Donations=pd.DataFrame(Notmapped_Donations)
Donations.columns=['Category','Contributor','Occupation','Date_of_Donation','Amount','Recepient']
#dropping duplicates
Donations.drop_duplicates(None,keep='first',inplace=True)
#dropping records with null amount 
Donations=Donations.dropna(subset=['Amount'])
Donations['Contributor']=Donations['Contributor'].str.replace("\n","")
#Droppig columns
Donations=Donations.drop(['Recepient','Category'],axis=1)

Donations['LastName']=Donations.Contributor.str.split('\s+').str[2]
Donations['LastName']=Donations['LastName'].str.replace(",","")

Donations['FirstName']= Donations.Contributor.str.split('\s+').str[3]
Donations['FirstName']=Donations['FirstName'].str.replace(",","")
Donations['Year']=Donations['Date_of_Donation'].astype(str).str.split('-').str[2]
Donations=Donations[['LastName','FirstName','Year','Amount']]
Donations['Amount']=Donations['Amount'].str.replace("$","")
Donations['Amount']=Donations['Amount'].str.replace(",","")
Donations['Amount']=pd.to_numeric(Donations['Amount'])
Donations=Donations.groupby(['LastName','FirstName','Year'])['Amount'].sum()
Donations_fin=Donations.reset_index()
NotmappedDonations_final=Donations_fin
```
